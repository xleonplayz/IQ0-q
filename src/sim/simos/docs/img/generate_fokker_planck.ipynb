```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Parameters
no_pathes = 3
pts = 100

# Generate random walk trajectories
y_ax = np.linspace(0, 1, pts)
all_pathes = np.zeros((no_pathes, pts))

np.random.seed(3)
for i in range(no_pathes):
    x_ax = np.ones(pts) * i
    freqs = [1, 2, 3, 4, 5]
    coeffs = np.random.normal(0, 1, len(freqs)) / 10
    phases = np.random.uniform(0, 2 * np.pi, len(freqs))
    for j, freq in enumerate(freqs):
        x_ax += coeffs[j] * np.sin(2 * np.pi * freq * y_ax + phases[j])
    all_pathes[i] = x_ax

# Create the figure and axis
fig, ax = plt.subplots()
ax.set_xlim(-1, no_pathes)
ax.set_ylim(0, 1)

# Plot the paths as lines
for i in range(no_pathes):
    ax.plot(all_pathes[i], y_ax, 'k-', alpha=0.5)

# Initialize the dots
colors = ['#BD1622', '#F29100', '#009EE2']
dot_objs = [ax.plot([], [], 'o', color=colors[i],ms=10)[0] for i in range(no_pathes)]

# grey-filled square boxes around the labels 
for i in range(no_pathes):
    ax.add_patch(plt.Rectangle((i-0.5, 1.02), 1, 0.05, fc='lightgrey', ec='k'))


# add Hamiltonian label over the pathes
for i in range(no_pathes):
    ax.text(i, 1.05, '$\mathcal{H}_' + str(i) + '$', fontsize=12, ha='center', va='bottom', color='k')


# Remove the axes
ax.set_xticks([])
ax.set_yticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Update function for animation
def update(frame):
    x_data = all_pathes[:, frame]
    y_data = y_ax[frame]
    for i in range(no_pathes):
        dot_objs[i].set_data(x_data[i], y_data)
    return dot_objs

# Create animation
ani = FuncAnimation(fig, update, frames=pts, blit=True, repeat=True)

# Save as GIF
ani.save('random_walk.gif', writer=PillowWriter(fps=10))

plt.show()
```
